```python
# Cleaning the jsonl file
import json
import re

# File paths
input_file = r"C:\Users\justu\OneDrive\Documents\Data_science\Portfolio\OHP LLM\chat_formatted_train.jsonl"  
output_file = r"C:\Users\justu\OneDrive\Documents\Data_science\Portfolio\OHP LLM\output_cleaned.jsonl"  

def clean_content(content):
    """
    Cleans the content by fixing encoding issues and removing placeholders.
    """
    # Fix encoding issues
    content = content.replace("\u00c2", "").replace("\u00c3", "").replace("\u00a1", "¡").replace("\u00bf", "¿")
    
    # Remove placeholders
    if "Placeholder completion" in content:
        return None
    
    # Remove unnecessary symbols like "->"
    content = re.sub(r"->", "", content).strip()
    return content

def reformat_chat_data(input_path, output_path):
    """
    Reformats the dataset to be compatible with chat fine-tuning.
    """
    cleaned_data = []
    
    with open(input_path, "r", encoding="utf-8") as infile:
        for line in infile:
            # Load each line as JSON
            try:
                entry = json.loads(line)
                messages = entry.get("messages", [])
                
                # Skip invalid entries
                if not messages or len(messages) < 2:
                    continue
                
                # Clean and validate messages
                reformatted_messages = []
                for message in messages:
                    content = clean_content(message.get("content", ""))
                    if content:
                        reformatted_messages.append({
                            "role": message.get("role", ""),
                            "content": content
                        })
                
                # Ensure all three roles are present (system, user, assistant)
                roles = [msg["role"] for msg in reformatted_messages]
                if "system" in roles and "user" in roles and "assistant" in roles:
                    cleaned_data.append({"messages": reformatted_messages})
            except json.JSONDecodeError:
                continue
    
    # Write cleaned data to output file
    with open(output_path, "w", encoding="utf-8") as outfile:
        for entry in cleaned_data:
            json.dump(entry, outfile)
            outfile.write("\n")
    print(f"Cleaned and reformatted data saved to {output_path}")

# Run the function
reformat_chat_data(input_file, output_file)
```
